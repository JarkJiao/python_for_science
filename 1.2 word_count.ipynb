## Word count(version 1)

```python
#speech
dic = {}
for word in speech:
    if word not in dic:
        dic[word] = 1
    else:
        dic[word] += 1
```

```python
#dic
```

```python
import operator
swd = sorted(dic.items(),key = operator.itemgetter(1),reverse=True)
```

```python
#swd
```

```python
from nltk.corpus import stopwords
stop_words = stopwords.words('English')
```

```python
for k,v in swd:
    if k not in stop_words:
        print (k,v)
```
